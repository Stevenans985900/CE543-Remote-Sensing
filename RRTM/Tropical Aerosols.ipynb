```python
names = ['level', 'pressure', 'UpwardFlux', 'DifDownFlux', 'DirDownFlux', 'DownwardFlux', 'NetFlux', 'HeatingRate']
aod0_00 = pd.DataFrame(loadtxt(ds[0]+'OUTPUT_RRTM'), columns=names)
aod0_25 = pd.DataFrame(loadtxt(ds[1]+'OUTPUT_RRTM'), columns=names)
aod0_50 = pd.DataFrame(loadtxt(ds[2]+'OUTPUT_RRTM'), columns=names)
aod0_75 = pd.DataFrame(loadtxt(ds[3]+'OUTPUT_RRTM'), columns=names)
aod1_00 = pd.DataFrame(loadtxt(ds[4]+'OUTPUT_RRTM'), columns=names)
```

```python
figure(figsize=(12,8))
plot(aod0_00.UpwardFlux, aod0_00.pressure,
     aod0_25.UpwardFlux, aod0_25.pressure,
     aod0_50.UpwardFlux, aod0_50.pressure,
     aod0_75.UpwardFlux, aod0_75.pressure,
     aod1_00.UpwardFlux, aod1_00.pressure)
gca().invert_yaxis()
xlabel('Flux (W m-2)')
ylabel('Pressure (mb)')
title('Upward Flux')
legend(['0.0', '0.25', '0.5', '0.75', '1.0'], loc='best')
```
