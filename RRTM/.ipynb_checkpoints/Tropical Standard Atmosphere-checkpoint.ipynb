```python
%pylab
import pandas as pd
```

```python
trp = pd.read_table('trp.dat', sep='\s+', names=['layer', 'altitude', 'pressure', 'temperature', 'dew_point', 'density', 'h2o', 'co2', 'o3', 'n2o', 'co', 'ch4', 'o2'])
```

```python
plt.figure(figsize=(12,12))
plt.semilogy(trp.temperature-273.15, trp.pressure)
plt.gca().invert_yaxis()
xlabel('Temperature (C)')
ylabel('Pressure (mb)')
title('Tropical Standard Atmosphere')
```

```python
plt.figure(figsize=(12,12))
plt.semilogy(trp.h2o/1e6*100, trp.pressure)
plt.gca().invert_yaxis()
xlabel('Water Vapor (%)')
ylabel('Pressure (mb)')
title('Tropical Standard Atmosphere')
```

```python
plt.figure(figsize=(12,12))
plt.semilogy((trp.co2.values).astype(int), trp.pressure)
plt.gca().invert_yaxis()
xlabel('Carbon Dioxide (ppm)')
ylabel('Pressure (mb)')
title('Tropical Standard Atmosphere')
```
