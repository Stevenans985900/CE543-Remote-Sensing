# Comparison of the Doppler and Lorentz line widths

### This comparison is computed as a function of pressure (and frequency) using Equation 9.37 in Petty (2006), A first course in Atmospheric Radiation.

```python
%pylab inline
```

```python
# Set pressure and temperature.
p  = 300.
T  = 230.               # Kelvin

# Doppler line
v0 = 31.26e12            # 1042 cm-1 in Hz (for Ozone).
kB = 1.3806488e-23       # Boltzmann constant
m  = (3.*16) / 6.02e23   # Molar mass divided by Avogadro's number
c2 = (3e8)**2            # Speed of light squared.
aD = v0 * (2.*kB*T/(m*c2))

varr = arange(-0.2,0.2,0.001)
fD = 1./(aD*pi**0.5)*exp(-varr**2 / aD**2)

# Lorentz line
p0 = 1013.5                      # mb
T0 = 273.15                      # K
a0 = 0.05                         # Typical value for a0
n  = 0.6                         # Typical value for n (I think)

aL = a0 * (p/p0) * (T0/T)**n
#varr = arange(-2,2,0.001)
fL = (aL/pi)/(varr**2 + aL**2)

plot(varr/aD,fD,varr/aL,fL)
axis([-4,4,0,max(fD)*1.1])
xlabel('(v-v0)/aD')
ylabel('Normalized f(v-v0)')
title('Doppler and Lorentz Line Shapes')
legend(('Doppler','Lorentz'),loc='best')
```

```python
# Equation 9.37 from Petty(2006).
v0      = 20.01e12                    # v0 must be in Hz.
p       = arange(1., 1000., 1.)     # p must be in mb.
a_ratio = 5.e-13 * v0 / p
```

```python
loglog(a_ratio,p)
xlabel('aD / aL')
ylabel('pressure (mb)')
title('Ratio of the Doppler and Lorentz line widths')
```
